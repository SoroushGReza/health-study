```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Stil för grafeerna
sns.set()

np.random.seed(42)
```

```python
df = pd.read_csv("../data/health_study_dataset.csv")

df.head()
```

```python
cols = ["age", "weight", "height", "systolic_bp", "cholesterol"]

descriptive_stats = df[cols].agg(["mean", "median", "min", "max"])
descriptive_stats
```

## Beskrivande statistik

Tabellen ovan visar medelvärde, median, min och max för ålder, vikt, längd,
systolistiskt blodtryck och kolesterol. Detta ger första överblicket över
hälsoprofilen i studien.
